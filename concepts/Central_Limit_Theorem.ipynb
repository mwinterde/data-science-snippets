# Central Limit Theorem

Another important concept of inferential statistics is the Central Limit Theorem. It says that if $X$ has a distribution with mean $\mu$ and standard deviation $\sigma$, and if $X$ is normally distributed or $n$ is large, then the distribution of its sample means $\bar{x}$ is approximately normally distributed with mean $\mu$ and standard deviation $\frac{\sigma}{\sqrt{n}}$.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

In the following I will first show that this is true for a normally distributed population. You can clearly see that the distribution of sample means $\bar{x}$ is again normally distributed and centered around the true population mean $\mu$. This distribution gets narrower as the sample size increases. By the way, this also suggests why we get smaller confidence intervals for larger samples. We can just be more 'confident' that the observed sample mean is nearby the true population mean.

```python
# Set parameters
population_mean = 100
population_std = 10
sample_size_ls = [5, 20, 50]

# Plot distribution of sample means for various sample sizes
fig, axes = plt.subplots(1, 3, figsize=(20, 5), sharex=True)
for i, sample_size in enumerate(sample_size_ls):
    # Draw 10000 samples of size n from population
    samples = np.random.normal(population_mean, population_std, size=(10000, sample_size))
    
    # Calculate sample means
    sample_means = np.mean(samples, axis=1)
    
    # Plot distribution of sample means
    axes[i].set_title("n = "+str(sample_size))
    axes[i].set_xlabel("Sample Mean")
    axes[i].hist(sample_means)
```

The actual power of the Central Limit Theorem lies in the fact, that these claims are true for all distributions. For illustration, I perform the exact same procedure as before, but know with a population that follows an exponential distribution. And what you can see is that as $n$ increases the distribution of the sample mean $\bar{x}$ converges to a normal distribution.

```python
# Set parameters
population_std = 100
sample_size_ls = [5, 20, 50]

# Plot distribution of sample means for various sample sizes
fig, axes = plt.subplots(1, 3, figsize=(20, 5), sharex=True)
for i, sample_size in enumerate(sample_size_ls):
    # Draw 10000 samples of size n from population
    samples = np.random.exponential(population_std, size=(100000, sample_size))
    
    # Calculate sample means
    sample_means = np.mean(samples, axis=1)
    
    # Plot distribution of sample means
    axes[i].set_title("n = "+str(sample_size))
    axes[i].set_xlabel("Sample Mean")
    axes[i].hist(sample_means, bins=30)
```
